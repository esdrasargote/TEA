```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Datos_TEA.csv")
data
```

```python
CO2 = data[data.Elemento == "Emisiones (CO2)"]
CO2
```

```python
Co2 = CO2.groupby(["Área","Año"], sort=False)["Valor"].max()
Co2.unstack().T.plot(kind="bar", subplots=False, layout=(2,2),figsize=(20,10), width=1, color=("red","orange","yellow","pink"))
```
